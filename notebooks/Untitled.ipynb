```python
grouped = df_US.groupby('title')[['trending_date']].count().sort_values('trending_date', ascending=False).head(10).copy()
grouped['trending_date']
```

```python
titles = list(grouped.index)
# titles
views = list(grouped['trending_date'].values)
# views
titles = titles[::-1]
views = views[::-1]
titles
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.barh(titles, views)
plt.xlabel("Number of days trending")
plt.ylabel('Videos')
plt.title("Num days trending for top 10 videos")
plt.savefig('top trending videos')
```
